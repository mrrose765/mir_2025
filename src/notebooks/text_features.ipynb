```python
from sentence_transformers import SentenceTransformer
import json
from tqdm import tqdm
import os
```

# 1 - Load dataset

```python
with open("captions_MIR_DATASETS_B.json") as f:
    data = json.load(f)
```

# 2 - Load model


```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

# 3 - Load features

```python
text_features = {}

for image_name, description in tqdm(data.items()):
    embedding = model.encode(description, convert_to_tensor=True)
    text_features[image_name] = embedding.tolist()
print("Text features encoded")
```

# 4 - Save features

```python
for image_name, embedding in tqdm(text_features.items()):

    filename = image_name.strip().split("/")[-1]
    filename = filename.replace('.jpg', '.txt')

    # Create folder text features
    os.makedirs("text_features", exist_ok=True)
    # Save features
    # Save embedding
    with open(f"text_features/{filename}", "w") as f:
        for i in range(len(embedding)):
            f.write(f"{embedding[i]}\n")

```
